# Data Context
The datasets which will be used in this thesis are the Adult Income dataset and the Student Performance dataset.
The Adult Income dataset is from the UC Irvine Machine Learning Repository and holds data from the census in the United States of America in 1994. The purpose of a census is to collect data from citizens to help decision making in e.g., construction, transportation and education.



# Data Description of Adult Income
The Adult Income Dataset has 48842 entries and 14 attributes. The attributes are:
- age, which is a continuous variable indicating the age of the individual
- workclass, which is a categorical variable indicating the working status of individual. The possible values are: Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay and Never-worked.
- fnlwgt, which is a continuous variable indicating how many similar people this individual represents
- education, which is a categorical variable indicating the highest level of diploma the individual possesses. The possible values are: Preschool, 1st-4th, 5th-6th, 7th-8th, 9th, 10th, 11th, 12th, HS-grad, Some-college, Bachelors, Masters, Prof-school, Assoc-acdm, Assoc-voc and Doctorate.
- educational-num, which is a continuous/categorical variable indicating 'education' numerical
- marital-status, which is a categorical variable indicating the marital status of the individual. The possible values are: Married-civ-spouse, Married-AF-spouse, Married-spouse-absent, Divorced, Never-married, Separated and Widowed.
- occupation, which is a categorical variable indicating profession of the individual. The possible values are: Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-speciality, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farm-fishing, Transport-moving, Priv-house-serv, Protective-serv and Armed-Forces.
- relationship, which is a categorical variable indicating what the relation is which this individual is to others. The possible values are: Wife, Own-child, Husband, Not-in-family, Other-relative and Unmarried.
- race, which is a categorical variable indicating the ethnicity of the individual. The possible values are: White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other and Black.
- gender, which is a categorical variable indicating the gender of the individual. The possible values are: Female and Male.
- capital-gain, which is a continuous variable indicating the capital gain of the individual
- capital-loss, which is a continuous variable indicating the capital loss of the individual
- hours-per-week, which is a continuous variable indicating the amount of hours the individual works
- native-country, which is a categorical variable indicating the country of birth/origin of the individual. The possible values are: United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinidad&Tobago, Peru and Hong.
- income, which is a categorical variable indicating whether the individual has an income above \$50,000 or not annually
